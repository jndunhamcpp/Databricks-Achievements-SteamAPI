```python
import time
import requests
from src.steam.client import SteamClient
from src.config.settings import (
    STEAM_API_KEY,
    STEAM_BASE_URL,
)

POPULAR_GAMES = {
    730: "CS2",
    570: "Dota 2",
    440: "Team Fortress 2",
    578080: "PUBG",
    271590: "GTA V",
    1085660: "Destiny 2",
    1172470: "Apex Legends",
    1091500: "Cyberpunk 2077",
}

steam_client = SteamClient(
    base_url="https://api.steampowered.com",
    api_key="YOUR_API_KEY",
)

def test_app(appid: int, name: str):
    print(f"\n=== Testing {name} ({appid}) ===")

    # ---- Player count ----
    try:
        count = steam_client.get_number_of_current_players(appid)
        print(f"Player count OK: {count}")
    except requests.HTTPError as e:
        print(f"Player count FAILED: {e.response.status_code}")
    except Exception as e:
        print(f"Player count ERROR: {e}")

    time.sleep(1.5)

    # ---- Achievements ----
    try:
        achievements = steam_client.get_global_achievements(appid)
        print(f"Achievements OK: {len(achievements)} achievements")
    except requests.HTTPError as e:
        print(f"Achievements FAILED: {e.response.status_code}")
    except Exception as e:
        print(f"Achievements ERROR: {e}")

    time.sleep(2.0)


if __name__ == "__main__":
    for appid, name in POPULAR_GAMES.items():
        test_app(appid, name)
```
